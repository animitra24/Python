```python
df2 = pd.read_csv('C:\\Users\\User\\Desktop\\Anindita Mitra\\Machine Learning\\EDA\\Fruits_Sales_MV_Python.csv')
df2
```

```python
# axis 1,0 (0- columns default fill data column by columns | 1 fill data row by row)
df2.interpolate()
```

```python
# limit : to fill particular number 
df2.interpolate(limit = 1)
```

```python
# limit 2 : maximum upto 2 limit of null value will be filled per column 
df2.interpolate(limit = 2)
```

```python
# fill value using backward method
df2.interpolate(limit =1 , limit_direction = 'backward')
```

```python
# fill value using backward method
df2.interpolate(limit =2 , limit_direction = 'backward')
```

```python
# both fill value using backward method (1 value forward and 1 value backward 2 consiquitve nulll value)
df2.interpolate(limit =1 , limit_direction = 'both')
```

```python
 # Area indide the data value will be filled
df2.interpolate(limit_area = 'inside')
```

```python
df2.interpolate(limit_area = 'outside')
```
