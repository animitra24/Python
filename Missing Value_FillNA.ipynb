# Fill NA

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\User\Desktop\Anindita Mitra\Machine Learning\EDA\Fruits_Sales_MV_Python.csv")
df
```

```python
# filling NAN value with 0
df.fillna(0)
```

```python
# Value : filling NAN value with 2
df.fillna(2)
```

```python
# filling NAN value with seperate value for each column
df.fillna({'ITEMSSOLD':27,'SALES':5000,'Profit':3000 })
```

```python
#Method ffill : forwrd fill. It helps to fill the value using previous value
df.fillna(method='ffill')
```

```python
#Method bfill : backward fill. It helps to fill the value using next value
#last NAN is not fill because there is no value after that Row-8
df.fillna(method='bfill')
```

```python
#axis : None (needs either value/method to execute the command axis)
#axis : NAN will be filled with next columns value
df.fillna(method='bfill',axis = 0)
```

```python
#axis : None (needs either value/method to execute the command axis)
#axis : NAN will be filled with previous columns value
df.fillna(method='bfill',axis = 1)
```

```python
#limit : specific NAN value is filled
df.fillna(0, limit=1)
```

```python
#inplace 
df.fillna(0, inplace=True)
df
```
