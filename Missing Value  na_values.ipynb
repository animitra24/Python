# keep_default_na

```python
# keep_default_na = False : this will keep the default value for na as per orginal csv file
df = pd.read_csv("Fruits_Sales_MV_Python.csv", keep_default_na = False)
df
```

```python
 # keep_default_na = True : this will keep the ignore the default value and consider this as NAN
df = pd.read_csv("Fruits_Sales_MV_Python.csv", keep_default_na = True)
df
```
