# Pandas Merge Function

Pandas merge connects columns or indexes in DataFrame based on one or more keys

```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:\\Users\\User\\Desktop\\Anindita Mitra\\Machine Learning\\EDA\\Cafe_Sales.csv')
df1
```

```python
df2 = pd.read_csv('C:\\Users\\User\\Desktop\\Anindita Mitra\\Machine Learning\\EDA\\Cafe_Region.csv')
df2
```

```python
#Intersection : Only common value is merrged not the rest
pd.merge(df1,df2, on ='PRODUCTLINE')
```

```python
pd.merge(df2,df1, on ='PRODUCTLINE')
```

```python
pd.merge(df1,df2, how ='inner')
```

```python
pd.merge(df2,df1, how ='left')
```

```python
pd.merge(df2,df1, how ='right',indicator = True)
```

```python
pd.merge(df2,df1, how ='outer')
```

```python
pd.merge(df2,df1, how ='outer', indicator = True)
```
