## Python Pivot Table

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\User\Desktop\Anindita Mitra\Machine Learning\EDA\Cafe_Year.csv')
```

```python
df.head(10)
```

```python
# values : aggregate by Product Name
df.pivot_table(index='Product Name')
```

```python
#lets aggregate by Year
3# If any value is not present for that year it will reflect as 'NAN'
df.pivot_table(index='Product Name', columns = 'Year')
```

```python
#aggfunc'mean'
df.pivot_table(index='Product Name', columns = 'Year', aggfunc = 'mean')
```

```python
df.pivot_table(index='Product Name', columns = 'Year', aggfunc = 'max')
```

```python
df.pivot_table(index='Product Name', columns = 'Year', aggfunc = 'min')
```

```python
df.pivot_table(index='Product Name', columns = 'Year', aggfunc = 'sum')
```

```python
#Margings : sum and then average as per attribute 
df.pivot_table(index='Product Name', columns = 'Year', aggfunc = 'sum', margins=True)
```

```python
#Margings : sum and then average as per attribute 
df.pivot_table(index='Country', columns = 'Year', aggfunc = 'sum', fill_value = 0, margins=True)
```

```python
#Margings : sum and then average as per attribute 
df.pivot_table(index='Product Name', columns = 'Year',  margins=True)
```
