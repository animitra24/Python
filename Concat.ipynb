# Concat

```python
import pandas as pd
```

```python
sr1 = pd.Series([0,1,2])
```

```python
sr2 = pd.Series([3,4,5,6,7])
```

```python
#obs (equal series)
pd.concat([sr1,sr2])
```

```python
#obs (unequal series)
pd.concat([sr1,sr2])
```

```python
df1 = pd.read_csv('C:\\Users\\User\\Desktop\\Anindita Mitra\\Machine Learning\\EDA\\Fruits_Profit.csv')
```

```python
df2 = pd.read_csv('C:\\Users\\User\\Desktop\\Anindita Mitra\\Machine Learning\\EDA\\Cafe_Region.csv')
```

```python
#axis = 0 (rows)
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
#Columns wise
pd.concat([df2,df1], axis =1)
```

```python
pd.concat([df1,df2], axis =1, ignore_index=True)
```

```python
#print common values for both the values we will use join
pd.concat([df1,df2], axis =1, join='inner')
```

```python
pd.concat([df1,df2], keys=['df1', 'df2'])
```

```python
pd.concat([df1,df2], axis=1,  keys=['df1', 'df2'])
```
